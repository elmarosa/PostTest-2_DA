```python
df2 = df.drop(['Played By', 'Played With', 'Played Vs.'], axis=1)
df2
```

```python
df2[['T_WinRate', 'T_PickPercentage', 'BS_WinRate', 'RS_WinRate', 'T_BansPercentage', 'T_PicksBansPercentage']] = df[['T_WinRate', 'T_PickPercentage', 'BS_WinRate', 'RS_WinRate', 'T_BansPercentage', 'T_PicksBansPercentage']].replace('%', '', regex=True)
df2
```

```python
df2[['T_WinRate', 'T_PickPercentage', 'BS_WinRate', 'RS_WinRate', 'T_BansPercentage', 'T_PicksBansPercentage']] = df2[['T_WinRate', 'T_PickPercentage', 'BS_WinRate', 'RS_WinRate', 'T_BansPercentage', 'T_PicksBansPercentage']].replace('-', 0, regex=True)
df2
```

```python
df2[['T_WinRate', 'T_PickPercentage', 'BS_WinRate', 'RS_WinRate', 'T_BansPercentage', 'T_PicksBansPercentage']] = df2[['T_WinRate', 'T_PickPercentage', 'BS_WinRate', 'RS_WinRate', 'T_BansPercentage', 'T_PicksBansPercentage']].astype(float)
df2
```

```python
df2.info()
```

```python
df2[['T_WinRate', 'T_PickPercentage', 'BS_WinRate', 'RS_WinRate', 'T_BansPercentage', 'T_PicksBansPercentage']] = df2[['T_WinRate', 'T_PickPercentage', 'BS_WinRate', 'RS_WinRate', 'T_BansPercentage', 'T_PicksBansPercentage']].astype(str) + '%'
df2
```

```python
sns.set()
```

```python
# Distribution of Win Rates

sns.set_theme(style="whitegrid")
sns.displot(df2["T_WinRate"], kde=True, color="thistle", height=6, aspect=2)
plt.title("Distribution of Win Rates", fontweight="bold", fontsize=15)
plt.xlabel("Win Rate")
plt.ylabel("Count")
plt.tight_layout()
plt.show();
```

```python
# Win Rate Distribution each Hero

df3 = df2.head(20)

plt.figure(figsize=(25, 6))
sns.barplot(x='Hero', y='T_WinRate', data=df3, color="Thistle")
plt.title('Win Rate Distribution for Each Hero', fontweight="bold", fontsize=25)
plt.xlabel('Hero')
plt.ylabel('Win Rate')
plt.show()
```

```python
# Comparison of Win Rate Between Blue Side and Red Side

sns.set_theme(style="whitegrid")
sns.barplot(y=['Blue Side', 'Red Side'], x=[df2['BS_WinRate'].mean(), df2['RS_WinRate'].mean()], palette=["steelblue", "firebrick"])
plt.title('Comparison of Win Rate Between Blue Side and Red Side', fontweight='bold', fontsize=10)
plt.xlabel('Average Win Rate (%)')
plt.ylabel('Side Team')
plt.show();
```

```python
# Distribution of Hero Roles // COUNT

plt.figure(figsize=(12, 6))
sns.set_theme(style="whitegrid")
sns.countplot(df2, x="Roles", color='Thistle')
plt.title("Distribution of Hero Roles", fontweight='bold', fontsize=15)
plt.xlabel('Hero Roles')
plt.ylabel('Count')
plt.xticks(rotation=90);
```

```python
# Average Win Rate by Hero Role
average_win_rate_by_role = df2.groupby('Roles')['T_WinRate'].mean()

plt.figure(figsize=(12, 6))
sns.barplot(x=average_win_rate_by_role.index, y=average_win_rate_by_role.values, color="Thistle")
plt.title('Average Win Rate by Hero Roles', fontweight='bold', fontsize=15)
plt.xlabel('Hero Roles')
plt.ylabel('Average Win Rate')
plt.xticks(rotation=90)
plt.show();
```

```python
# Correlation between Hero Picks & Win Rate

plt.figure(figsize=(15, 6))
sns.scatterplot(x='T_PickPercentage', y='T_WinRate', data=df2, color='orchid')
plt.title('Correlation between Hero Pick Percentage and Win Rate', fontweight='bold', fontsize=15)
plt.xlabel('Hero Pick Percentage (%)')
plt.ylabel('Win Rate')
plt.show()
```

```python
# Ban Percentage for Each Hero

df3 = df2.head(20)

plt.figure(figsize=(20, 6))
sns.barplot(x=df3['Hero'], y=df3['T_BansPercentage'], color='Thistle')
plt.title('Ban Percentage for Each Hero', fontweight='bold', fontsize=22)
plt.xlabel('Hero')
plt.ylabel('Ban Percentage (%)')
plt.show()
```

```python
# High Pick and Ban Hero Percentages

df3 = df2.head(20)

pick_percentage = df3['T_PickPercentage']
ban_percentage = df3['T_BansPercentage']
heroes = df3['Hero']

plt.figure(figsize=(20, 6))
plt.bar(heroes, pick_percentage, color='Thistle', label='PICK')
plt.bar(heroes, ban_percentage, bottom=pick_percentage, color='plum', label='BAN')
plt.title('High Pick and Ban Hero Percentages', fontweight='bold', fontsize=22)
plt.xlabel('Hero')
plt.ylabel('Percentage (%)')

plt.legend()
plt.tight_layout()
plt.show();
```

```python
# Relationship between Pick, Bans, Wins

plt.figure(figsize=(20, 8))
sns.set_theme(style="whitegrid")
sns.scatterplot(data=df2, x='T_Picked', y='T_Banned', size='T_Wins', hue='T_Wins', palette=sns.dark_palette('pink'), sizes=(30, 300))
plt.title('Relationship between Pick, Bans, and Wins', fontweight='bold', fontsize=22)
plt.xlabel('Number of Picks')
plt.ylabel('Number of Bans')

plt.legend(title='Win Rate', loc='upper left', bbox_to_anchor=(1, 1), ncol=3)
plt.tight_layout()
plt.show();
```
